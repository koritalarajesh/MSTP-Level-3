## Random Forest
- collection of decision trees is called a Random forest
- It plays a key role for 1 lakh column of dataset
- Random forest is applicable both classification and regression


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("diabates.csv")
data
```

```python
data.columns
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
X= data.drop('Outcome',axis=1)
Y= data['Outcome']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.30,random_state=1)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(X_train,Y_train)
y_pred=rf.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(Y_test,y_pred))
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy=accuracy_score(Y_test,y_pred)
```

# Classification Report
- precision: what percentage of your preditions were correct
      - precision= TP/(TP+FP)
- recall: what percentage of positive cases you got
      - Recall= TP/(TP+FP)
- F1 score: what percentage of positive predictions were correct 
      - F1 score= (2*(Recall* precision))/(Recall+precision)

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test,y_pred))
```
